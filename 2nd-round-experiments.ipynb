```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl

platesem2 = np.load('AGURP2019Emission.npy') 
platesab2 = np.load('AGURP2019Absorption.npy') 

wavelengths_em_i4 = range(380,602,2)
wavelengths_ab_i4 = range(250,702,2)

times_h_i1 = [0.5*(i+1) for i in range(11)] + [11*0.5 + 1.5*(i+1) for i in range(53-11)] 
# the number of hours after addition that the sample was measured
# I2 --> the second index corresponds to your samples in rows from A-H. It is not uniform across columns, so you cannot generalize. 
# Refer to the Experiment Preparation sheet
# I3 --> the third index corresponds to your samples in columns from 1-12. It is not uniform across rows, so you cannot generalize.

```

```python
print(times_h_i1)
```

```python
plt.figure(figsize=(8,8))
for i in range(30,len(platesab2)):
    plt.plot(wavelengths_ab_i4,
             platesab2[i][5][6],
             label=str(times_h_i1[i]) + ' hours',
                      linewidth = 3)

plt.xlim(250,530)
plt.ylim(0,3)
plt.title('Evolution of Absorption in Well A2')
plt.ylabel('Absorptivity')
plt.xlabel('Wavelength (nm)')
plt.legend()
plt.show()
```
